```python
%%bash

rm -rf /tmp/wmt16_en_de
mkdir -p /tmp/wmt16_en_de
```

```python
%%bash

# Download Europarl v7
wget -nc -nv -O /tmp/wmt16_en_de/training-parallel-europarl-v7.tgz \
  http://www.statmt.org/wmt13/training-parallel-europarl-v7.tgz
```

```python
%%bash

# Download Common Crawl
wget -nc -nv -O /tmp/wmt16_en_de/training-parallel-commoncrawl.tgz \
  http://www.statmt.org/wmt13/training-parallel-commoncrawl.tgz
```

```python
%%bash

# Download News 11
wget -nc -nv -O /tmp/wmt16_en_de/training-parallel-nc-v11.tgz \
  http://data.statmt.org/wmt16/translation-task/training-parallel-nc-v11.tgz
```

```python
%%bash

# Download test and dev sets
wget -nc -nv -O /tmp/wmt16_en_de/dev.tgz \
  http://data.statmt.org/wmt16/translation-task/dev.tgz
wget -nc -nv -O /tmp/wmt16_en_de/test.tgz \
  http://data.statmt.org/wmt16/translation-task/test.tgz
```

```python
%%bash

# Extract everything
mkdir -p /tmp/wmt16_en_de/train
tar -xvzf /tmp/wmt16_en_de/training-parallel-europarl-v7.tgz -C /tmp/wmt16_en_de/train
tar -xvzf /tmp/wmt16_en_de/training-parallel-commoncrawl.tgz -C /tmp/wmt16_en_de/train
tar -xvzf /tmp/wmt16_en_de/training-parallel-nc-v11.tgz -C /tmp/wmt16_en_de/train
tar -xvzf /tmp/wmt16_en_de/dev.tgz -C /tmp/wmt16_en_de
```

```python
%%bash

# Download MOSES for preprocessing
wget -nv -nc -O /tmp/wmt16_en_de/moses.zip \
  https://github.com/moses-smt/mosesdecoder/archive/master.zip
unzip -d /tmp/wmt16_en_de /tmp/wmt16_en_de/moses.zip
```

```python
%%bash

# Concatenate EN Training data
cat /tmp/wmt16_en_de/train/training/europarl-v7.de-en.en \
  /tmp/wmt16_en_de/train/commoncrawl.de-en.en \
  /tmp/wmt16_en_de/train/training-parallel-nc-v11/news-commentary-v11.de-en.en \
  > /tmp/wmt16_en_de/train/combined-de-en.en
  
wc -l /tmp/wmt16_en_de/train/combined-de-en.en
```

```python
%%bash

# Concatenate DE Training data
cat /tmp/wmt16_en_de/train/training/europarl-v7.de-en.de \
  /tmp/wmt16_en_de/train/commoncrawl.de-en.de \
  /tmp/wmt16_en_de/train/training-parallel-nc-v11/news-commentary-v11.de-en.de \
  > /tmp/wmt16_en_de/train/combined-de-en.de
  
wc -l /tmp/wmt16_en_de/train/combined-de-en.de  
```

```python
%%bash

# Convert newstest2015 data into raw text format
/tmp/wmt16_en_de/mosesdecoder-master/scripts/ems/support/input-from-sgm.perl \
  < /tmp/wmt16_en_de/dev/newstest2015-ende-src.en.sgm \
  > /tmp/wmt16_en_de/dev/newstest2015.en

# Convert newstest2015 data into raw text format
/tmp/wmt16_en_de/mosesdecoder-master/scripts/ems/support/input-from-sgm.perl \
  < /tmp/wmt16_en_de/dev/newstest2015-ende-ref.de.sgm \
  > /tmp/wmt16_en_de/dev/newstest2015.de
```

```python
%%bash

# Tokenization directory
mkdir -p /tmp/wmt16_en_de/tokenized

# Tokenize newstest 2015
/tmp/wmt16_en_de/mosesdecoder-master/scripts/tokenizer/tokenizer.perl -l en -threads 4 \
  < /tmp/wmt16_en_de/dev/newstest2015.en \
  > /tmp/wmt16_en_de/tokenized/newstest2015.en.tok
/tmp/wmt16_en_de/mosesdecoder-master/scripts/tokenizer/tokenizer.perl -l de -threads 4 \
  < /tmp/wmt16_en_de/dev/newstest2015.de \
  > /tmp/wmt16_en_de/tokenized/newstest2015.de.tok  
  
# Tokenize newstest 2013
/tmp/wmt16_en_de/mosesdecoder-master/scripts/tokenizer/tokenizer.perl -l en -threads 4 \
  < /tmp/wmt16_en_de/dev/newstest2013.en \
  > /tmp/wmt16_en_de/tokenized/newstest2013.en.tok
/tmp/wmt16_en_de/mosesdecoder-master/scripts/tokenizer/tokenizer.perl -l de -threads 4 \
  < /tmp/wmt16_en_de/dev/newstest2013.de \
  > /tmp/wmt16_en_de/tokenized/newstest2013.de.tok
```

```python
%%bash
wc -l /tmp/wmt16_en_de/tokenized/newstest2013.en.tok
wc -l /tmp/wmt16_en_de/tokenized/newstest2013.de.tok
wc -l /tmp/wmt16_en_de/tokenized/newstest2015.en.tok
wc -l /tmp/wmt16_en_de/tokenized/newstest2015.de.tok
```

```python
%%bash

# Tokenize combined training data
/tmp/wmt16_en_de/mosesdecoder-master/scripts/tokenizer/tokenizer.perl -l en -threads 4 \
  < /tmp/wmt16_en_de/train/combined-de-en.en \
  > /tmp/wmt16_en_de/tokenized/combined-de-en.en.tok
  
/tmp/wmt16_en_de/mosesdecoder-master/scripts/tokenizer/tokenizer.perl -l de -threads 4 \
  < /tmp/wmt16_en_de/train/combined-de-en.de \
  > /tmp/wmt16_en_de/tokenized/combined-de-en.de.tok
```

```python
%%bash

# Create EN Vocabulary
../seq2seq/scripts/generate_vocab.py \
  --input_file /tmp/wmt16_en_de/tokenized/combined-de-en.en.tok \
  --output_file /tmp/wmt16_en_de/tokenized/vocab.en \
  --max_vocab_size 50000
```

```python
%%bash

# Create DE Vocabulary
../seq2seq/scripts/generate_vocab.py \
  --input_file /tmp/wmt16_en_de/tokenized/combined-de-en.de.tok \
  --output_file /tmp/wmt16_en_de/tokenized/vocab.de \
  --max_vocab_size 50000
```

```python
%%bash

# Convert parallel training data to TFRecords (this may take a while)
../seq2seq/scripts/generate_examples.py \
  --source_file /tmp/wmt16_en_de/tokenized/combined-de-en.en.tok \
  --target_file /tmp/wmt16_en_de/tokenized/combined-de-en.de.tok \
  --output_file /tmp/wmt16_en_de/tokenized/train.tfrecords
```

```python
%%bash

# Convert parallel newstest2013
../seq2seq/scripts/generate_examples.py \
  --source_file /tmp/wmt16_en_de/tokenized/newstest2013.en.tok \
  --target_file /tmp/wmt16_en_de/tokenized/newstest2013.de.tok \
  --output_file /tmp/wmt16_en_de/tokenized/newstest2013.tfrecords
  
  # Convert parallel newstest2015
../seq2seq/scripts/generate_examples.py \
  --source_file /tmp/wmt16_en_de/tokenized/newstest2015.en.tok \
  --target_file /tmp/wmt16_en_de/tokenized/newstest2015.de.tok \
  --output_file /tmp/wmt16_en_de/tokenized/newstest2015.tfrecords  
```
